```python
# to check for all duplicate values
df = df.drop_duplicates().reset_index(drop=True)
df
```

```python
# for missing values
num_cols = df.select_dtypes(include=['number']).columns.tolist()
cat_cols = df.select_dtypes(exclude=['number']).columns.tolist()

# median for numeric
for c in num_cols:
    df[c] = df[c].fillna(df[c].median())

# mode for categorical (fall back to 'Unknown' if all NaN)
for c in cat_cols:
    if df[c].isnull().any():
        mode = df[c].mode(dropna=True)
        if len(mode)>0:
            df[c] = df[c].fillna(mode[0])
        else:
            df[c] = df[c].fillna('Unknown')

df
```

```python
# Removing outliers using IQR method for numeric columns
# Removing rows outside IQR bounds (default)
def remove_outliers_iqr(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5*iqr
    upper = q3 + 1.5*iqr
    return df[(df[col] >= lower) & (df[col] <= upper)]

for c in num_cols:
    df = remove_outliers_iqr(df, c)
```

```python
# Categorical cleaning & encoding 
for c in cat_cols:
    counts = df[c].value_counts(normalize=True)
    rare = counts[counts < 0.01].index
    df[c] = df[c].replace(rare, 'Other')
```

```python
# Inspect data types of each column
print(df.head())
print(df.dtypes)
```

```python
# inspecting data types of each column and checking for dummies
country_cols = [c for c in df.columns if c.startswith("Country_")]

if country_cols:
    # Create 'Country' column by checking which dummy is True
    df["Country"] = df[country_cols].idxmax(axis=1).str.replace("Country_", "")
    
    # Drop dummy columns (optional, keep only if you need them for ML)
    df = df.drop(columns=country_cols)

#  Final checks

print("Final dataset shape:", df.shape)
print(df.dtypes)
print(df.head())

#  Save cleaned dataset

output_csv   = r"C:\Users\Lenovo\Downloads\climate_change_dataset_cleaned.csv"
output_excel = r"C:\Users\Lenovo\Downloads\climate_change_dataset_cleaned.xlsx"

df.to_csv(output_csv, index=False)
df.to_excel(output_excel, index=False)

print(f"\n Cleaned dataset with 'Country' restored saved to:\n- {output_csv}\n- {output_excel}")
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Select features for risk score
features = df[['Avg Temperature (Â°C)', 'Extreme Weather Events', 'Renewable Energy (%)']]

# Scale them between 0 and 1
scaled = pd.DataFrame(scaler.fit_transform(features), columns=features.columns)

# Custom Climate Risk Score
# Higher Renewable Energy should reduce risk â subtract it
df['Risk Score'] = (
    scaled['Avg Temperature (Â°C)'] * 0.3 +
    scaled['Extreme Weather Events'] * 0.4 +
    (1 - scaled['Renewable Energy (%)']) * 0.3
)

# Average-risk per country
risk = df.groupby('Country')['Risk Score'].mean().sort_values(ascending=False)

# Plot bar chart
risk.plot(kind='bar', figsize=(10,5), color='orangered')
plt.title("Climate Risk Score by Country (Normalized Index)")
plt.ylabel("Risk Score (0-1)")
plt.show()
```
